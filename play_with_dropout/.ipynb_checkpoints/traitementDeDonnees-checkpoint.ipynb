```python
"""
les symboles particulières comme ? ! et . sont supprimé dans la question.
Il faut peut être returner une réponse spécifique si il rencontre plus 1 mot qu'il existe pas dans le corpus
Nous suprimmons également:
  1. Mot qui contient une seule lettre(qui sont suivante des erreurs)
  2. Une fonction unique pour nettoyage tous les données dans train-test et chat
Nous allons traiter tous les mots clés suivant :
km, vitesse, carburant, huile, position, batterie, pression_pneu. Il reste trajet et vin
  
"""
#import sys
#!{sys.executable} -m pip uninstall gtts

#from gtts import gTTS
#from playsound import playsound
#import speech_recognition
from nltk.corpus import stopwords
STOPWORDS = stopwords.words('french')
from nltk.stem import *
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("french", ignore_stopwords = False)
import csv
import re
from __future__ import unicode_literals, print_function, division
from io import open
import unicodedata
import string
import re
import random
import os
import math
import time
from dateparser.search import search_dates
import pandas as pd


    
def clearing_word(word):
    word = re.sub('\x8e', 'é', word)
    word = re.sub('\x88', 'à', word)
    word = re.sub('\x9d', 'ù', word)
    word = re.sub('\x8f', 'è', word)
    word = re.sub('\x9e', 'û', word)
    word = re.sub('\x90', 'ê', word)
    word = re.sub('\x99', 'ô', word)
    word = re.sub('\x94', 'î', word)
   # word = re.sub('\x8f', 'è', word)
    word = re.sub('\x8d', 'ç', word)
    word = re.sub('õ', '', word)
    word = re.sub('Ê', '', word)
    word = re.sub('[?,.,!, \,,  %]', '', word)
    if word == 'û' or word == 'v' or word == 'é':
        word = ''
    if word =="2017êles":
        word = "2017"
    if "ênox" in word:
        word ="nox"
    return word


def clear_line(pharse, stop_word):
    """
    Arg: string
    Return: string
    """
    dict_manuel = dict_manuelle()
    pharse_ =[]
    for words in pharse.strip().strip('?').strip('.').strip('!').lower().split(' '):
        for word in words.split('-'):
            for word_ in word.split('\''):
                word_ = clearing_word(word_)
                word_ = stemmer.stem(word_)
        if word_ not in stop_word and len(word_) >1:       
            pharse_.append(word_)  
        for i in range(len(pharse_)):
            for line in dict_manuel:
                if pharse_[i] in line:
                    pharse_[i] = line[0]
    return ' '.join(pharse_)


def get_all_convos(stopwords, file):
    
    convos = []
    #file = 'questions13juillet.txt'
    with open(file) as f:
            i=0
            for line in f:
                if '#' in str(line):
                    #print(line)
                    line = line.strip()
                    question_to_hold = str(line)
                else:
                    if len(line)==1 and question_to_hold != 'no':
                        convos.append(question_to_hold)
                    else:
                        question_to_hold = 'no'
                    
                i+=1
    convos_to_return = []
    index_of_good = []
    for pair in convos:
        convos_to_return.append(clear_line(pair, stopwords))
    convos2 = []
    index = 0
    for pair in convos_to_return:
        if pair not in convos2:
            convos2.append(pair)
            index_of_good.append(index)
        index +=1
    good_convos = [convos[i] for i in index_of_good]
    print('len convos', len(convos))
    print(len(convos), '  ', len(good_convos))
    #save_convos_in_file(good_convos)
    
def get_all_and_all_questions():
    file1 = 'questions16juillet.txt'
    file2 = 'questions13juillet.txt'
    files = [file1, file2]
    convos = []
    lines = []
    for file in files:
        with open(file) as f:
            for line in f:
                if '#' in line:
                    line = line.strip().strip('#').strip('?').strip('!')
                    if len(lines) >0:
                        convos.append(lines)
                    lines = [line]
                else:
                    if len(line) > 1:
                        line = line.strip().strip('#').strip('?').strip('!')
                        lines.append(line)
        f.close()
    return convos

def get_all_convos_in_train_test_file():
    
    convos = []
    file1 = 'convos27juin_test_smaller.txt'
    file2 = 'convos27juin_train_smaller.txt'
    #file1 = 'chatbot_tout_corpus.txt'
    #file2 = 'chatbot_tout_corpus_13juin.txt'
    #file3 = 'convos_120_nettoye.txt' # cette fichier viens d'etre ajouter
    liste_file = [file1, file2]
    list_of_word = []
    for file in liste_file:
        with open(file) as f:
            i=0
            for line in f:
                if i%2==0:
                    question = str(line)
                    if '++++' in question:
                         question = question[9:]
                    question = question.strip().strip('?').strip('#').strip('!')
                else:
                    answer = str(line)
                    if '++++' in answer:
                        answer = answer[9:].strip()
                    convos.append([question, answer])
                i+=1
        convos_to_return = []
        for pair in convos:
            if len(pair[0]) !=0:
                convos_to_return.append(pair)
        f.close()
    return convos_to_return

def clear_pairs(pairs):
    pairs_return =[]
    for pair in pairs:
        pairs_return.append([clear_line(pair[0], STOPWORDS), pair[1]])
    return pairs_return

def find_good_pair_of_index(all_question, all_pairs):
    number = 0
    good_index = []
    not_good_questions = []
    pairs = []
    for i in range(len(all_question)):
        boolean = False
        for question in all_question[i]:
            for j in range(len(all_pairs)):
                if question == all_pairs[j][0]:
                    answer_index = j
                    boolean = True
                    break
        if boolean:
            good_index.append([i, answer_index])
    print('len of all question: {} and len of good_index:{}, len of all_pairs: {}'.\
             format(len(all_question), len(good_index), len(all_pairs)))
    return good_index
 


def process_raw_data_and_return_good_convos():
    
    raw_questions = get_all_and_all_questions()
    all_pairs = get_all_convos_in_train_test_file()
    all_pairs_clean = clear_pairs(all_pairs)
    raw_questions_clean = []
    for questions in raw_questions:
        question_clean = []
        for question in questions:
            question_clean.append(clear_line(question, STOPWORDS))
        raw_questions_clean.append(question_clean)
    good_index = find_good_pair_of_index(raw_questions_clean, all_pairs_clean)
    print('len of raw question {} and len of good questions {}'. format(len(raw_questions), len(good_index)))
    good_pairs = []
    for indexs in good_index:
        for question in raw_questions[indexs[0]]:
                good_pairs.append([question, all_pairs[indexs[1]][1]])
    good_pairs.extend(all_pairs)
    good_pairs_ = delete_all_pairs_duplicate_after_stemmer_and_dict(good_pairs)
    index_of_train = random.sample(range(len(good_pairs_)), int(.80*len(good_pairs_)))
    print('len of train {} and len of test {}'.format(len(index_of_train), len(good_pairs_)-len(index_of_train)))
    with open('trains_16juillet.txt', 'w') as f:
        for index in index_of_train:
            if '\n' in good_pairs_[index][0]:
                question = 'YOU++++ ' +str(good_pairs_[index][0])
            else:               
                question = 'YOU++++ ' +str(good_pairs_[index][0])+'\n'
            if '\n' in good_pairs_[index][1]:
                answer = 'BOT++++ '+ str(good_pairs_[index][1])
            else:
                answer = 'BOT++++ '+ str(good_pairs_[index][1])+'\n'
            f.write(question)
            f.write(answer)
    f.close()
    with open('tests_16juillet.txt', 'w') as f:
        for index in range(len(good_pairs_)):
            if index not in index_of_train:
                question = 'YOU++++ ' +str(good_pairs_[index][0])+'\n'
                answer = 'BOT++++ '+ str(good_pairs_[index][1])+'\n'
                f.write(question)
                f.write(answer)
    f.close()
   # return good_pairs_
        
                
                    
        

def delete_all_pairs_duplicate_after_stemmer_and_dict(all_pairs):
    
    dict_ = dict_manuelle()
    questions_stemmer = [clear_line(all_pairs[i][0], STOPWORDS) for i in range(len(all_pairs))]
    questions_by_dict = []
    for question in questions_stemmer:
        words = []
        for word in question.split():
            if word in dict_:
                words.append(dict_[word][0])
            else:
                if len(word)>0:
                    words.append(word)
        words = ' '.join(words)
        questions_by_dict.append(words)       
    index = []
    questions_ = []
    for i in range(len(questions_by_dict)):
        if  questions_by_dict[i] not in questions_:
            index.append(i)
            questions_.append(questions_by_dict[i])
    return [all_pairs[i] for i in index]


def dict_manuelle():
    """
    return la liste des listes des synonymes"""
    list_ = []
    dict_ = {}
    dict_['voiture'] = ['voiture', 'véhicule']
    dict_['km'] = ['km', 'kilomètrage', 'kilométrage', 'distance', 'kilomètre']
    dict_['essence'] = ['essence', 'carburant', 'énergie']
    dict_['roule'] = ['roule', 'roulent', 'marche', 'marcher', 'circule', 'circulent', 'circulation']
    dict_['bien'] = ['bien', 'bonne', 'good', 'excellent', 'parfait', 'beau', 'ok']
    dict_['min']  = ['min', 'minimum', 'minimal']
    dict_['pneu'] = ['pneumatique', 'pneu', 'pneus']
    dict_['pouquoi'] = ['cause', 'pourquoi', 'raison', 'motif']
    dict_['stupid'] = ['stupid', 'con', 'idiot', 'bête', 'imbécile']
    dict_['connaître'] = ['connaître', 'connaitre', 'savoir', 'comprendre',\
                          'connaisance', 'connais', 'connaîs']
    dict_['problème']  = ['problème', 'difficulté', 'danger', 'ennui', 'dangers']
    dict_['erreur'] = ['faute', 'erreur']
    dict_['arrêt'] = ['arrêt', 'arrêter', 'stopper', 'immobilisé', 'immobiliser', 'immobile', 'paralisé']
    dict_['disponible'] = ['disponible', 'disponibilité', 'libre']
    dict_['abimer'] = ['abîmer', 'hors service', 'cassé', 'endommager', 'abîmé']
    dict_['boitier'] = ['boitier', 'boîtier', 'boîte']
    dict_['frein'] = ['frein', 'freinage', 'freiner']
    dict_['parcouru'] = ['parcouru', 'parcourus', 'parcourir']
    dict_['peux'] = ['peux', 'pouvoir', 'pourrais']
    dict_['veux'] = ['veux', 'vouloir', 'veut', 'voulais', 'voudrais', 'souhaiter', 'souhaite']
    dict_['fait'] = ['fait', 'fais', 'faire', 'faites']
    dict_['fort'] = ['fort', 'puissant', 'robuste', 'solide']
    dict_['grand'] = ['grand', 'gross', 'élevé', 'haut', 'haute']
    dict_['mal'] = ['mal', 'mauvais', 'souci']
    dict_['changer'] = ['modifier', 'modification', 'changer', 'changement']
    dict_['avoir'] = ['as', 'obtenir', 'avoir', 'posséder', 'disposer']
    dict_['peur']  = ['peur', 'craint']
    dict_['placer'] = ['placer','installer']
    dict_['nombre'] = ['nombre', 'combien', 'quantité']
    dict_['aider']  = ['aider', 'aide', 'soutien', 'soutenir']
    dict_['question'] = ['question', 'demande']
    dict_['information'] = ['information', 'infos', 'renseignement', 'indictation',\
                            'informer', 'renseigner','indiquer']
    dict_['usé'] = ['usé', 'usure', 'vieux', 'vieille', 'fatigué']
    dict_['réduire']  = ['réduire', 'réduction', 'abaisser', 'diminuer', 'diminution']
    dict_['présent']  = ['présent', 'maitenant', 'actuel', 'actuellement']
    dict_['suivre']   = ['suivre', 'observer', 'surveiller', 'poursuivre']
    dict_['aller']    = ['aller', 'vas', 'vais', 'va']
    dict_['échanger'] = ['échanger', 'estimer', 'échangement', 'estimation',\
                         'évaluer', 'calculer','évaluation'\
                        , 'mesurer']
    dict_['taux']    = ['taux', 'pourcentage']
    dict_['ampoule'] = ['ampoule', 'feux', 'feu']
    for key in dict_:
        l = []
        for word in dict_[key]:
            l.append(stemmer.stem(word))
        list_.append(l)
    return list_


def save_convos_in_file(all_convos):
    with open('questions_13juillet_netoye.txt', 'w') as f:
        for pair in all_convos:
                f.write('#'+pair+'\n')
    f.close()
   
```
