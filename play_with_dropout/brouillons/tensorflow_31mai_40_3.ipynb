```python
from __future__ import division
from __future__ import print_function

import argparse
import os
os.environ['TF_CPP_MIN_LOG_LEVEL']='2'
import random
import sys
import time
import math

import numpy as np
import tensorflow as tf

from model_40_3 import ChatBotModel
import config_tf as config
import data

import matplotlib.pyplot as plt

def _get_random_bucket(train_buckets_scale):
    rand = random.random()
    return min([i for i in range(len(train_buckets_scale))
                if train_buckets_scale[i] > rand])


def _assert_lengths(encoder_size, decoder_size, 
                    encoder_inputs, decoder_inputs, decoder_masks):

    if len(encoder_inputs) != encoder_size:
        raise ValueError("Encoder length must be equal to the one in bucket,"
                        " %d != %d." % (len(encoder_inputs), encoder_size))
    if len(decoder_inputs) != decoder_size:
        raise ValueError("Decoder length must be equal to the one in bucket,"
                       " %d != %d." % (len(decoder_inputs), decoder_size))
    if len(decoder_masks) != decoder_size:
        raise ValueError("Weights length must be equal to the one in bucket,"
                       " %d != %d." % (len(decoder_masks), decoder_size))

def run_step(sess, model, encoder_inputs, decoder_inputs,
             decoder_masks, bucket_id, forward_only):
    encoder_size, decoder_size = config.BUCKETS[bucket_id]
    input_feed = {}
    for step in range(encoder_size):
        input_feed[model.encoder_inputs[step].name] = encoder_inputs[step]
    for step in range(decoder_size):
        input_feed[model.decoder_inputs[step].name] = decoder_inputs[step]
        input_feed[model.decoder_masks[step].name] = decoder_masks[step]
    
    last_target = model.decoder_inputs[decoder_size].name
    input_feed[last_target] = np.zeros([model.batch_size], dtype=np.int32)
    if not forward_only:
        output_feed = [model.train_ops[bucket_id],  # update op that does SGD.
                       model.gradient_norms[bucket_id],  # gradient norm.
                       model.losses[bucket_id]]  # loss for this batch.
    else:
        output_feed = [model.losses[bucket_id]]  # loss for this batch.
        for step in range(decoder_size):  # output logits.
            output_feed.append(model.outputs[bucket_id][step])
    outputs = sess.run(output_feed, input_feed)
    if not forward_only:
        return outputs[1], outputs[2], None  # Gradient norm, loss, no outputs.
    else:
        return None, outputs[0], outputs[1:]  # No gradient norm, loss, outputs.

def _get_buckets():
    all_data = data.load_data('question2id.txt', 'answer2id.txt')
    all_data = all_data[0]
    total_len = len(all_data)
    portion_of_train = int(total_len*0.85)
    index_of_train = random.sample(range(total_len), portion_of_train)
    data_train, data_test = [], [] 
    for index in range(total_len):
        if index in index_of_train:
            data_train.append(all_data[index])
        else:
            data_test.append(all_data[index])
    
    return  data_train, data_test

def _get_skip_step(n_iters):
    """ How many steps should the model train before it saves all the weights. """

    return int(n_iters/10)

def _check_restore_parameters(sess, saver):
    """ Restore the previously trained parameters if there are any. """
    ckpt = tf.train.get_checkpoint_state(os.path.dirname(config.CPT_PATH_40_3 + '/checkpoint'))
    if ckpt and ckpt.model_checkpoint_path:
        print("Loading parameters for the Chatbot")
        saver.restore(sess, ckpt.model_checkpoint_path)
    else:
        print("Initializing fresh parameters for the Chatbot")

def _eval_test_set(sess, model, test_buckets):
    """ Evaluate on the test set. """
    for bucket_id in range(len(config.BUCKETS)):
        if len(test_buckets[bucket_id]) == 0:
            print("  Test: empty bucket %d" % (bucket_id))
            continue
        start = time.time()
        encoder_inputs, decoder_inputs, decoder_masks = data.get_batch(test_buckets[bucket_id], 
                                                                        bucket_id,
                                                                        batch_size=config.BATCH_SIZE)
        _, step_loss, _ = run_step(sess, model, encoder_inputs, decoder_inputs, 
                                   decoder_masks, bucket_id, True)
        print('Test bucket {}: loss {}, time {}'.format(bucket_id, step_loss, time.time() - start))



def asMinutes(s):
    m = math.floor(s / 60)
    s -= m * 60
    return '%dm %ds' % (m, s)


def timeSince(since, percent):
    now = time.time()
    s = now - since
    es = s / (percent)
    rs = es - s
    return '%s (- %s)' % (asMinutes(s), asMinutes(rs))

def train(n_iters):
    """ Train the bot """
    data_buckets = data_train
    tf.reset_default_graph() 
    model = ChatBotModel(False, config.BATCH_SIZE)
    model.build_graph()

    saver = tf.train.Saver()

    with tf.Session() as sess:
        print('Running session')
        sess.run(tf.global_variables_initializer())
        _check_restore_parameters(sess, saver)

        iteration = model.global_step.eval()
        total_loss = 0
        step =0
        loss_print = []
        bucket_id = 0
        start = time.time()
        skip_step = _get_skip_step(n_iters)
        while step < n_iters:
            step+=1
            encoder_inputs, decoder_inputs, decoder_masks = data.get_batch(data_buckets, 
                                                                           bucket_id,
                                                                           batch_size=config.BATCH_SIZE)
            
            _, step_loss, _ = run_step(sess, model, encoder_inputs, 
                                       decoder_inputs, decoder_masks, bucket_id, False)
            total_loss += step_loss
            iteration += 1
            if iteration % skip_step == 0:  
                print_loss_avg = total_loss / skip_step
                total_loss = 0
                print('%s (%d %d%%) %.4f' % (timeSince(start, step / n_iters),
                                         iteration, step / n_iters * 100, print_loss_avg))
                total_loss = 0
                loss_print.append(step_loss)
        saver.save(sess, os.path.join(config.CPT_PATH_40_3, 'chatbot'), global_step=model.global_step)
        sys.stdout.flush()
        plt.plot(loss_print)
        plt.show()

def _get_user_input():
    t = input('Vous:  ')
    return t

def _construct_reponse(output_logits, inv_dec_vocab):
    outputs = [int(np.argmax(logit, axis=1)) for logit in output_logits]
    if config.EOS_ID in outputs:
        outputs = outputs[:outputs.index(config.EOS_ID)]

    return " ".join([tf.compat.as_str(inv_dec_vocab[output]) for output in outputs])
def id2word(ids, inv_dec_vocab):
    return " ".join([tf.compat.as_str(inv_dec_vocab[output]) for output in ids[:-1]])

def chat():
    _, enc_vocab = data.load_vocab(os.path.join(config.PROCESSED_PATH, 'encoder_vocab.txt'))
    inv_dec_vocab, _= data.load_vocab(os.path.join(config.PROCESSED_PATH, 'decoder_vocab.txt'))
    tf.reset_default_graph() 
    model = ChatBotModel(True, batch_size=1)
    model.build_graph()

    saver = tf.train.Saver()

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        _check_restore_parameters(sess, saver)
        output_file = open(os.path.join(config.PROCESSED_PATH, config.OUTPUT_FILE), 'a+')
        max_length = config.BUCKETS[-1][0]
        print('Bonjour, dites moi ce que vous voulez: ')
        while True:
            line = _get_user_input()
            if len(line) > 0 and line[-1] == '\n':
                line = line[:-1]
            if line == '':
                break
            output_file.write('VOUS ++++ ' + line + '\n')
            token_ids = data.sentence2id(enc_vocab, str(line))
            if (len(token_ids) > max_length):
                print('La longueur maximale est:', max_length)
                continue
            bucket_id = _find_right_bucket(len(token_ids))
            encoder_inputs, decoder_inputs, decoder_masks = data.get_batch([(token_ids, [])], 
                                                                            bucket_id,
                                                                            batch_size=1)
            _, _, output_logits = run_step(sess, model, encoder_inputs, decoder_inputs,
                                           decoder_masks, bucket_id, True)
            response = _construct_reponse(output_logits, inv_dec_vocab)
            
            print('Bot de AVICEN:  ', response)
            output_file.write('BOT ++++ ' + response + '\n')
        output_file.write('=============================================\n')
        output_file.close()
        
def evaluate_randomly(n_iters, test=True):
    enc_vocab, _ = data.load_vocab(os.path.join(config.PROCESSED_PATH, 'encoder_vocab.txt'))
    inv_dec_vocab, _= data.load_vocab(os.path.join(config.PROCESSED_PATH, 'decoder_vocab.txt'))
    if not test:
        encode_file = open(os.path.join(config.PROCESSED_PATH, "question2id.txt"), 'r')
        decode_file = open(os.path.join(config.PROCESSED_PATH, "answer2id.txt"), 'r')
        encode, decode = encode_file.readline(), decode_file.readline()
        id_questions = []
        id_answers = []
        while encode and decode:
            encode_id = [int(id_) for id_ in encode.split()]
            decode_id = [int(id_) for id_ in decode.split()]
            id_questions.append(encode_id)
            id_answers.append(decode_id)
            encode, decode = encode_file.readline(), decode_file.readline()
            
        tf.reset_default_graph() 
        model = ChatBotModel(True, batch_size=1)
        model.build_graph()

        saver = tf.train.Saver()

        with tf.Session() as sess:
            sess.run(tf.global_variables_initializer())
            _check_restore_parameters(sess, saver)
            for i in range(n_iters):
                random_index = random.choice(range(len(id_questions)))
                bucket_id = 0
            
            
                encoder_inputs, decoder_inputs, decoder_masks = data.get_batch([(id_questions[random_index], [])], 
                                                                            bucket_id,
                                                                            batch_size=1)
                _, _, output_logits = run_step(sess, model, encoder_inputs, decoder_inputs,
                                           decoder_masks, bucket_id, True)
                reponse = _construct_reponse(output_logits, inv_dec_vocab)
                bonne_reponse = " ".join([str(inv_dec_vocab[id]) for id in id_answers[random_index][1:-1]])
                question = " ".join([str(enc_vocab[id]) for id in id_questions[random_index]])
                print('--------------------------------------------------')
                print('Question:  ',  question)
                print('Bot     :  ', reponse)
                print('Réponse : ', bonne_reponse)
    else:
        tf.reset_default_graph() 
        model = ChatBotModel(True, batch_size=1)
        model.build_graph()

        saver = tf.train.Saver()
        total_loss = 0

        with tf.Session() as sess:
            sess.run(tf.global_variables_initializer())
            _check_restore_parameters(sess, saver)
            n_iters = len(data_test)
            bucket_id = 0
            for i in range(n_iters):
                question = data_test[i][0]
                answer = data_test[i][1]
                encoder_inputs, decoder_inputs, decoder_masks = data.get_batch([(question, [])], 
                                                                            bucket_id,
                                                                            batch_size=1)
                _, _, output_logits = run_step(sess, model, encoder_inputs, decoder_inputs,
                                           decoder_masks, bucket_id, True)
                reponse = _construct_reponse(output_logits, inv_dec_vocab)
                bonne_reponse = " ".join([str(inv_dec_vocab[id]) for id in answer[1:-1]])
                loss = _evaluate_by_right_word(reponse, bonne_reponse)
                total_loss +=loss
                print('Reponse: ', reponse)
                print('Bonne Reponse: {}. ACCURACY {:.1f} '.format(bonne_reponse,1-loss))
                print('-'*50)
        print('Test on {} sentences'.format(n_iters))
        print('Accuracy by percent of true words {}'.format(1-total_loss/n_iters))
        #return loss/n_iters

def _evaluate_by_right_word(reponse, bonne_reponse):
    reponse = reponse.split()
    bonne_reponse = bonne_reponse.split()
    min_length = min(len(reponse), len(bonne_reponse))
    max_length = max(len(reponse), len(bonne_reponse))
    error = max_length-min_length
    for i in range(min_length):
        if reponse[i] != bonne_reponse[i]:
            error +=1
    return error/max_length

if __name__ =='__main__':
    try:
        data_train
    except NameError:
        data_train, data_test = _get_buckets()
    
```
